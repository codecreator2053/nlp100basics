## Problem 65

```python
from collections import defaultdict
lines =  open('questions-words.txt').readlines()
categories = defaultdict(list)
current_categ = ''

print(len(lines))
n = 0
for line in lines:
    n+=1
    if ': ' in line:
        current_categ = line[1:].strip()
    else:
        analogy = line.split(' ')
        predicted = model.most_similar(positive=[analogy[1], analogy[2]], negative=[analogy[0]], topn=1)[0][0]
        categories[current_categ].append([analogy[0], analogy[1], analogy[2], analogy[3].strip(), predicted])
        
```

```python
sem_cnt = 0
sem_cor = 0
syn_cnt = 0
syn_cor = 0
for i, j in categories.items():
    for k in j:
        if i.startswith('gram'):
            syn_cnt += 1
            if k[3] == k[4]:
                syn_cor +=1
        else:
            sem_cnt +=1
            if k[3] == k[4]:
                sem_cor +=1
                
print(f'semantic accuracy: {round(sem_cor/sem_cnt, 2)}')
print(f'syntactic accuracy: {round(syn_cor/syn_cnt, 2)}')
```

## Problem 67

```python
countries = open('countries.txt').readlines()
count_names = []
for i in countries:
    name = '_'.join(i.split('|')[1].strip().split(' '))
    count_names.append(name)
vectors = [model[name] for name in count_names if name in model]
names = [name for name in count_names if name in model]
```

```python
from sklearn.cluster import KMeans as kmeans
import numpy as np

vectors = np.array(vectors)
k = kmeans(n_clusters=7)
k.fit(vectors)
preds = k.predict(vectors)
df2 = pd.DataFrame(data = {'cluster': preds, 'names': names})
for i in range(7):
    print(f'cluster {i}')
    print(*list(df2[df2['cluster']==i]['names']), sep=', ')
```

```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
plt.figure(figsize=(20, 8))
Z = linkage(vectors, method='ward')
dendrogram(Z, labels=names)
plt.show()
```
